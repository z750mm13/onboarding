# Uso de funciones en Python

## Trabajar con argumentos en funciones

Los argumentos requeridos en las funciones se utilizan cuando las funciones necesitan que esos argumentos funcionen correctamente. En este ejercicio, construiremos un informe de combustible que requiere información de varias ubicaciones de combustible en todo el cohete.

Comenzaremos por crear una función que necesite tres lecturas de combustible y devuelva un informe:

```python
# Función para leer 3 tanques de combustible y muestre el promedio
def tanques(tanque1, tanque2, tanque3):
  average = (tanque1 + tanque2 + tanque3) / 3
  return f"Tank 1: {tanque1} Tank 2: {tanque2} Tank 3: {tanque3}. Average: {average}"
```

Ahora que hemos definido la función de informes, vamos a comprobarlo. Para esta misión, los tanques no están llenos:

```python
# Llamamos a la función que genera el reporte print(funcion(tanque1, tanque2, tanque3))
print(tanques(0, 0, 0))
```

En lugar de simplemente crear el informe, la función también está calculando el promedio. Mejoraremos la legibilidad extrayendo el cálculo promedio de la función en una nueva función para que el promedio se pueda hacer de forma independiente:

```python
# Función promedio
def promedio(valor1, valor2, valor3):
  return (valor1 + valor2 + valor3) / 3
```

Ahora actualizaremos la función de informes para llamar a la nueva función del promedio:

```python
# Actualiza la función
def tanques(tanque1, tanque2, tanque3):
  average = promedio(tanque1, tanque2, tanque3)
  return f"Tank 1: {tanque1} Tank 2: {tanque2} Tank 3: {tanque3}. Average: {average}"

print(tanques(2, 2, 2))
```
