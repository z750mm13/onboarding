## Uso de ciclos _while_ en Python

En Python, los ciclos _while_ te permiten ejecutar código un número desconocido de veces. Los ciclos examinan una condición booleana y, siempre que la condición sea verdadera, se ejecutará el código dentro del ciclo. Esto es muy útil para situaciones como solicitar valores a un usuario.

En este ejercicio, cremos una aplicación que solicite a un usuario que ingrese una lista de planetas. En un ejercicio posterior, agregaremos el código que muestre la lista. Por ahora, escribiremos solo el código que solicite al usuario la lista de planetas.

Comienzaremos agregando dos variables, una para la entrada del usuario, con el nombre new_planet, y otra variable para la lista de planetas, denominada planets.

```python
# Declara dos variables
planets = []
new_planet = ""
```

Comenzando con las variables que acabamos de crear, crearemos un ciclo while. El ciclo while se ejecutará mientras el new_planet no sea igual a la palabra 'done'.

Dentro del ciclo, comprobaremos si la variable new_planet contiene el valor que debería ser el nombre de un planeta. Esta es una forma rápida de ver si el usuario ha introducido un valor. Si lo han hecho, el código agregará (append) ese valor a la lista planets.

Finalmente, usaremos input para solicitar al usuario que ingrese un nuevo nombre de planeta o que escriba done si ha terminado de ingresar nombres de planeta. Almacenará el valor de input en la variable new_planet.

```python
# Escribe el ciclo while solicitado
while new_planet != "done":
  new_planet = input("Ingrese el nombre de un planeta o 'done' para finalizar: ")
  if new_planet != 'done':
    planets.append(new_planet)
```

## Creación de un ciclo "for"

En el ejercicio anterior, creamos el código para solicitar a los usuarios que introduzcan una lista de nombres de planetas. En este ejercicio, completaremos la aplicación escribiendo código que muestre los nombres de esos planetas.

La variable planets almacena los nombres de los planetas que ha introducido el usuario. Usaremos un ciclo para mostrar esas entradas.

Crearemos un ciclo for para iterar sobre la lista planets. Usaremos como nombre de la variable planet para cada planeta. Dentro del ciclo for.

```python
# Escribe tu ciclo for para iterar en una lista de planetas
for planet in planets:
  print(planet)
```
