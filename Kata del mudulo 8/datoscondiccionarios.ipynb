# Creación de diccionarios de Python

## Crear y modificar un diccionario de Python

Los diccionarios python permiten modelar datos más complejos. Los diccionarios son una colección de pares clave/valor, y son muy comunes en los programas Python. Su flexibilidad le permite trabajar dinámicamente con valores relacionados sin tener que crear clases u objetos.

Un diccionario se indica en Python mediante el uso de llaves ({ }), con pares clave/valor separados por dos puntos (:). Las claves son literales de cadena y los valores pueden ser de cualquier tipo de datos.{ }:

```python
demo = {
    "key": "value",
    "number": 42
}
```
Para este ejercicio, crearemos un diccionario que almacene información sobre el planeta Marte.

Agregaremos el código para crear un nuevo diccionario denominado 'planet'. Rellena con la siguiente información:

```
name: Mars
moons: 2
```

````python
# Crea un diccionario llamado planet con los datos propuestos
planet = {
    "name": "Mars",
    "moons": 2
}
````

Para recuperar valores, puede utilizar el método get o corchetes ([ ]) con el nombre de la clave que se desea recuperar.

````python
# Muestra el nombre del planeta y el número de lunas que tiene.
print("Planet name:", planet['name'], "Moons:", planet['moons'])
````

Podemos actualizar las claves existentes o crear otras nuevas utilizando el método update o entre corchetes ([ ]). Cuando se usa update, se pasa un nuevo objeto de diccionario con los valores actualizados o nuevos. Cuando se usan corchetes, se especifica el nombre de la clave y se asigna un nuevo valor.
```python
demo['new_key'] = 'New value'
```
Agregaremos un nuevo valor con una clave de 'circunferencia (km)'. Este nuevo valor almacenará un diccionario con las dos circunferencias del planeta:

```
polar: 6752
equatorial: 6792
```

````python
# Agrega la clave circunferencia con los datos proporcionados previamente
planet['circunferencia (km)'] = {
    'polar': 6752,
    'equatorial': 6792
}
````

Imprimiremos el nombre del planeta con su circunferencia polar.

````python
# Imprime el nombre del planeta con su circunferencia polar.
print(f"{planet['name']} has {planet['circunferencia (km)']['polar']} polar circumference")
````

## Programación dinámica con diccionarios

### Cálculo de valores

En este escenario, calcularemos tanto el número total de lunas en el sistema solar como el número promedio de lunas que tiene un planeta.

````python
# Planets and moons

planet_moons = {
    'mercury': 0,
    'venus': 0,
    'earth': 1,
    'mars': 2,
    'jupiter': 79,
    'saturn': 82,
    'uranus': 27,
    'neptune': 14,
    'pluto': 5,
    'haumea': 2,
    'makemake': 1,
    'eris': 1
}
````

Los diccionarios de Python permiten recuperar todos los valores y claves utilizando los métodos <code>values</code> y <code>keys</code>, respectivamente. Cada método devuelve una lista que contiene los datos, que luego se pueden usar como una lista normal de Python. Puedemos determinar el número de elementos mediante <code>len</code>, e iterar a través de él mediante un ciclo <code>for</code>.

Agregaremos continuación el código para determinar el número de lunas. Comenzaremos almacenando el valor <code>values</code> de <code>planet_moons</code> en una variable denominada <code>moons</code>. A continuación, almacenaremos el número de planetas en una variable denominada <code>planets</code>.

````python
# Añade el código para determinar el número de lunas.
moons = planet_moons.values()
planets = len(planet_moons.keys())
````

Agregaremos el código para contar el número de lunas. lo haremos creando un ciclo <code>for</code> para iterar a través de las lunas <code>moons</code> y agregándolos a una variable denominada <code>total_moons</code>. Finalmente calcularemos el promedio dividiendo <code>total_moons</code> por <code>planets</code> e imprimiendo los resultados.

````python
# Agrega el código para contar el número de lunas. 
total_moons = 0
for moon in moons:
  total_moons = total_moons + moon
average = total_moons / planets
print(f"The total moons is {total_moons} and the average is {average}.")
````
