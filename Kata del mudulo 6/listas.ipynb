# Creando y usando listas en Python

## Usar listas para almacenar nombres de planetas

Las listas permiten almacenar varios valores en una sola variable. Se comenzará un proyecto sobre información planetaria creando una lista de planetas.

En primer lugar, se creará una variable denominada planets. Agrega los ocho planetas (sin Plutón) a la lista. A continuación, muestra el número de planetas.

```python
# Creamos la lista planets y la mostramos
planetas = ['Mercurio', 'Venus', 'Tierra', 'Marte', 'Jupiter', 'Saturno', 'Urano', 'Neptuno']
```

Agregamos a Plutón a la lista que creamos. Luego mostramos tanto el número de planetas como el último planeta de la lista.

```python
# Agregamos a plutón y mostramos el último elemento
print("Se inició la lista con", len(planetas), "planetas.")

planetas.append("Plutón");
print("Ahora tenemos", len(planetas), " y el último elemento es", planetas[-1])
```

## Trabajando con datos de una lista

### Usar slices para recuperar partes de una lista

Es posible que se deba trabajar con diferentes secciones de una lista. En nuestro ejemplo, queremos mostrar planetas más cerca y más lejos del sol de un planeta que el usuario ingresa por teclado.

Comienzamos agregando el código para crear una lista con los planetas.

```python
# Lista de planetas
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Neptune']
```

A continuación solicitamos al usuario un nombre. Debido a que las cadenas distinguen entre mayúsculas y minúsculas en Python, pedimos al usuario que use una letra mayúscula para comenzar el nombre del planeta.

```python
# Solicitamos el nombre de un planeta *Pista:  input()*
planet = input("Ingrese el nombre de un planeta (Comenzando con la primera letra en mayuscula): ")
```

Para determinar qué planetas están más cerca que el que ingresó el usuario, se debe encontrar dónde está el planeta en la lista utilizando index para realizar esta operación.

```python
# Busca el planeta en la lista
user_planet = planets.index(planet)
```

Con el índice determinado, ahora podemos agregar el código para mostrar los planetas más cercanos al sol.

```python
# Muestra los planetas más cercanos al sol
print("Los planetas mas cercanos al Sol con respecto a", planet, "son", planets[0:user_planet])
```

Podemos usar el mismo índice para mostrar planetas más alejados del sol. Sin embargo, recuerdemos que el índice inicial se incluye cuando usamos un slice. Como resultado, tendremos que agregar 1 al valor.

```python
# Muestra los planetas más lejanos al sol
print("Los planetas mas lejanos al Sol con respecto a", planet, "son", planets[user_planet+1:])
```
