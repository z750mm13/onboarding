# Utilización de operadores aritméticos

### Operadores aritméticos en Python

Python proporciona operadores aritméticos comunes para que puedas realizar operaciones matemáticas en tu código. Estos incluyen las cuatro operaciones principales de suma, resta, multiplicación y división.

Exploremos cómo podemos crear un programa que pueda calcular la distancia entre dos planetas. Comenzaremos usando dos distancias de planetas: Tierra (149.597.870 km) y Júpiter (778.547.200 km).

```python
# Crear variables para almacenar las dos distancias
# ¡Asegúrate de quitar las comas!
distancia_de_la_tierra = 149597870
distancia_de_jupiter = 778547200
```

Con los valores obtenidos, es el momento de añadir el código para realizar la operación. Se restará el primer planeta del segundo para determinar la distancia en kilómetros. A continuación, vamos convertir la distancia de kilómetros a millas multiplicándola por 0.621

```python
# Calcular la distancia entre planetas
distancia_entre_planetas = abs(distancia_de_la_tierra - distancia_de_jupiter)
print("Distancia: "+ str(distancia_entre_planetas) +" kilometros")
distancia_entre_planetas  = round(distancia_entre_planetas * 0.621)
print("Distancia: "+ str(distancia_entre_planetas) +" millas")
```

### Convirtiendo cadenas en números y uso de valores absolutos

Creamos una aplicación para trabajar con números y entrada de usuario.

Con frecuencia, se deberán convertir los valores de las cadenas en números para realizar correctamente diferentes operaciones o determinar el valor absoluto de un número.

Para crear nuestra aplicación, leeremos la distancia del sol para dos planetas, y luego mostraremos la distancia entre los planetas. Haremos esto usando input para leer los valores, int para convertir a entero y luego abs para convertir el resultado en su valor absoluto.

```python
# Almacena las entradas del usuario
# Pista: variable = input("¿Cuál es tu nombre?")
primer_planeta = input("¿Cuál es la distancia del primer planeta?")
segundo_planeta = input("¿Cuál es la distancia del segundo planeta?")
```

Debido a que input devuelve valores de cadena, necesitamos convertirlos en números. Para nuestro ejemplo, usaremos int

```python
# Se convierten las cadenas de ambos planetas a números enteros
distancia_primer_planeta = int(primer_planeta)
distancia_segundo_planeta = int(segundo_planeta)
```

Con los valores almacenados como números, ahora podemos agregar el código para realizar el cálculo, restando el primer planeta del segundo. Debido a que el segundo planeta podría ser un número mayor, se usará abs para convertirlo a un valor absoluto. También se agregará el código para mostrar el resultado en millas multiplicando la distancia del kilómetro por 0.621

```python
# Se realiza el cálculo y se determinar el valor absoluto
distancia_entre_planetas = abs(distancia_primer_planeta - distancia_segundo_planeta)
print("Distancia: "+ str(distancia_entre_planetas) +" kilometros")
# Conversión de KM a Millas
distancia_entre_planetas  = round(distancia_entre_planetas * 0.621)
print("Distancia: "+ str(distancia_entre_planetas) +" millas")
```
